```r
High_data <- read.table('data_high.txt',header=T)
Low_data <- read.table('data_low.txt',header=T)
Neg_data <- read.table('data_neg.txt',header=T)
```

```r
High_data %>% group_by(sample) %>% summarise(sum = sum(end-start)/1000000) -> High_data_total_len
Low_data %>% group_by(sample) %>% summarise(sum = sum(end-start)/1000000) -> Low_data_total_len
Neg_data %>% group_by(sample) %>% summarise(sum = sum(end-start)/1000000) -> Neg_data_total_len
```

```r
High_number <- High_data$sample %>% table() %>% as.data.frame() %>% mutate(group="High")
```

```r
Low_number <- Low_data$sample %>% table() %>% as.data.frame() %>% mutate(group="Low")
```

```r
Neg_number <- Neg_data$sample %>% table() %>% as.data.frame() %>% mutate(group="Neg")
```

```r
High_data_total_len$group = "High"
Low_data_total_len$group = "Low"
Neg_data_total_len$group = "Neg"
```

```r
all_stat <- rbind(High_data_total_len,Low_data_total_len,Neg_data_total_len)[,2:3]
```

```r
number_stat <- rbind(High_number,Low_number,Neg_number)[,2:3]
```

```r
head(number_stat)
```

```r
compare = list(c("High","Low"),c("High","Neg"),c("Low","Neg"))
```

```r
p1 <- ggboxplot(all_stat,x="group",y="sum",add="jitter",add.params=list(shape=21, fill="orange", size=3))+
        geom_signif(comparisons = compare,step_increase = 0.1,map_signif_level = F,test = t.test)+
  ylab("Total length of CNA (/Mb)") + xlab("")
       # facet_wrap(~Genes, scale="free") 
```

```r
p2<- ggboxplot(number_stat,x="group",y="Freq",add="jitter",add.params=list(shape=21, fill="orange", size=3))+
        geom_signif(comparisons = compare,step_increase = 0.1,map_signif_level = F,test = t.test)+
  ylab("Total number of CNA") + xlab("")
       # facet_wrap(~Genes, scale="free") 
```

```r
p1
```

```r
p2
```

```r
ggsave(filename = "CNA_length_group.pdf",plot = p1,width = 4,height = 4)
```

```r
ggsave(filename = "CNA_number_group.pdf",plot = p2,width = 4,height = 4)
```
